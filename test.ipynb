```python
import yfinance as yf 
from techa.patterns import scan_patterns
```

```python
ohlcv = yf.download("STMMI.MI", start="2016-01-01", end="2026-04-28", multi_level_index=False)
```

```python
hits = scan_patterns(ohlcv)
```

```python
hits.tail(20)
```

```python
from techa.patterns import plot_pattern
plot_pattern(ohlcv, "CDLINVERTEDHAMMER", window_bars=10)
```

```python
tickers = ["STMMI.MI", "PRY.MI"] 
```

```python
ohlcv_by_ticker = {                                                                                                                                                
      t: yf.download(t, period="2y", multi_level_index=False)                                                                                                        
      for t in tickers                                                                                                                                               
  }
```

```python
from techa.patterns import scan_last_bar
report = scan_last_bar(ohlcv_by_ticker)
```

```python
report
```

```python
report = scan_last_bar(ohlcv_by_ticker, patterns=["CDLENGULFING", "CDLHAMMER", "CDLSHOOTINGSTAR"])
report
```

```python
from techa.patterns import scan_last_bar
from techa.indicators import build_snapshot
tickers = ["TEN.MI"]
ohlcv_by_ticker = {t: yf.download(t, period="1y", multi_level_index=False) for t in tickers}
patterns_report = scan_last_bar(ohlcv_by_ticker)
patterns_report
```

```python
ohlcv_by_ticker['STMMI.MI']
```

```python
build_snapshot(ohlcv_by_ticker['STMMI.MI'], nan_to_none=True)
```

```python
for ticker, ohlcv in ohlcv_by_ticker.items():
      snap = build_snapshot(ohlcv, nan_to_none=True)
      print(f"\n{ticker}")
      print(f"  ADX: {snap['adx']:.1f}  RSI: {snap['rsi']:.1f}  vol_vs_ma20: {snap['vol_vs_ma20']:.2f}")
      print(f"  dist_sma50: {snap['dist_sma50_pct']:.1f}%  golden_cross: {snap['golden_cross']}")
      fired = patterns_report[patterns_report["ticker"] == ticker]
      print(f"  Patterns: {fired['display_name'].tolist()}")
```
